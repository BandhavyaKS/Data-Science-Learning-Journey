## **Salary Prediction using Linear Regression**

**Objective**

Learn how Linear Regression predicts salary based on years of experience.

_________________________________________________________________________________________________________________________________

## **Step 1: Import Libraries**

We first import the required Python libraries. Pandas is used to handle the dataset, while Matplotlib is used for visualization.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## **Step 2: Load Dataset**

Read the CSV file into a DataFrame for further analysis.

```python
df=pd.read_csv('Salary_Data.csv' )
df
```

## **Step 3: Check Correlation**

Correlation helps us understand whether the independent variable and dependent variable are strongly related.

```python
df.corr()
```


Understanding:

Correlation measures the strength and direction of the relationship between the feature and the target variable. A correlation value close to +1 indicates a strong positive linear relationship.

Observation:

The correlation between YearsExperience and Salary is 0.978, which indicates a very strong positive relationship. Therefore, Linear Regression is a suitable model for this dataset.

## **Step 4: Scatter Plot**

A scatter plot helps verify whether the relationship is linear, which is an important assumption of Linear Regression.

```python
plt.scatter(df['YearsExperience'],df['Salary'] )# feature name# target name
```

A scatter plot is used to visualize the relationship between the feature (YearsExperience) and the target (Salary). It helps determine whether the data follows a linear pattern, which is an important assumption for Linear Regression.

Observation:

The data points show an upward linear trend, indicating that salary generally increases as years of experience increase. Since the relationship appears linear, Linear Regression is an appropriate model for this dataset.

## **Step 5: Build the Linear Regression Model**
Understanding

Linear Regression is a supervised machine learning algorithm used to predict a continuous target variable based on one or more input features.

What this code does
- Selects YearsExperience as the feature (X).
- Selects Salary as the target (y).
- Creates a LinearRegression model.
- Trains the model using the available data.

```python
from sklearn.linear_model import LinearRegression
X=df[['YearsExperience']]
Y=df['Salary']
model=LinearRegression()
model.fit(X,Y)
model
```

## **Step 6: Predict Salary Values**
Understanding

The predict() method uses the trained model to estimate salary values based on the given years of experience.

Observation

The predicted values represent the model's estimated salary for each input and can be compared with the actual salaries to evaluate performance.

```python
model.predict(X)
```

## **Step 8: Display Model Coefficient and Intercept**
Understanding

The coefficient represents the rate at which the salary changes with each additional year of experience, while the intercept represents the estimated salary when the years of experience is zero.

Observation

These values define the equation of the best-fit regression line used for prediction.

```python
print(model.coef_)
print(model.intercept_)
```

## **Step 9: Predict Salary for New Input**
Understanding

The trained model can also predict the salary for new experience values that were not part of the training dataset.

Observation

This demonstrates how Linear Regression can be used to make predictions on unseen data.

```python
model.predict([[5]])
```
